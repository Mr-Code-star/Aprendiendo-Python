```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("Adidas US Sales Datasets.xlsx") # Para leer el archivo Excel
df.columns = df.iloc[3] # Establecer la primera fila como encabezados de columna
df = df.iloc[4: ,1:].reset_index(drop=True) # Eliminar las filas anteriores y restablecer el 铆ndice
df['year'] = pd.to_datetime(df['Invoice Date']).dt.year # Extraer el a帽o de la columna 'Invoice Date'
df['month'] = pd.to_datetime(df['Invoice Date']).dt.month # Extraer el mes de la columna 'Invoice Date'
df['day'] = pd.to_datetime(df['Invoice Date']).dt.day # Extraer el d铆a de la columna 'Invoice Date'

```

```python
df.describe(include='all') # Para obtener estad铆sticas descriptivas de todas las columnas, incluyendo las no num茅ricas
```

```python
retalier_counts = df['Retailer'].value_counts() # Contar la cantidad de ventas por cada minorista
retalier_counts = retalier_counts.sort_values(ascending=False) # Ordenar los minoristas por cantidad de ventas
retalier_counts.plot(kind='bar') # Crear un gr谩fico de barras para visualizar la cantidad de ventas por minorista
plt.title('Cantidad de Ventas por Minorista') # T铆tulo del gr谩fico
plt.xlabel('Minorista') # Etiqueta del eje x
plt.ylabel('Cantidad de Ventas') # Etiqueta del eje y
plt.xticks(rotation=45) # Rotar las etiquetas del eje x para mejor legibilidad
plt.show() # Mostrar el gr谩fico
```

```python
import plotly.express as px

# Calcular las ventas totales por minorista
retailer_sales = df.groupby('Retailer')['Total Sales'].sum()


# Calcular la participaci贸n de mercado de cada minorista
total_sales = retailer_sales.sum()

market_share = retailer_sales / total_sales # Participaci贸n de mercado

# Crear un gr谩fico de pastel para visualizar la participaci贸n de mercado
fig = px.pie(market_share, values=market_share, names=market_share.index, title='Market Share of Retailers')

# Mostrar el gr谩fico
fig.show()
```

```python
# Covertir la columna 'Invoice Date' a formato datetime
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])

# Extraer el mes y el a帽o de la columna 'Invoice Date' y crear nuevas columnas para cada uno
df['month'] = df['Invoice Date'].dt.month
df['year'] = df['Invoice Date'].dt.year

# Crear una nueva columna 'year_month' combinando el a帽o y el mes para facilitar el an谩lisis temporal
df['year_month'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-01')

# Agrupar los datos por minorista, a帽o y mes, y calcular la suma de las ventas totales para cada grupo
retailer_sales = df.groupby(['Retailer', 'year_month'])['Total Sales'].sum().reset_index()

# Crear gr谩fico de l铆neas
fig = px.line(retailer_sales, 
              x='year_month', 
              y='Total Sales', 
              color='Retailer',
              title='Total Sales by Retailer Over Time',
              )

# Mostrar el gr谩fico
fig.show()
```

```python
product_sales = df.groupby(['Retailer', 'Product'])['Total Sales'].sum()

product_sales_df = product_sales.unstack(level=1)

product_sales_df.plot(kind='bar', title='Total Sales by Product and Retailer', figsize=(12, 6))

plt.show()
```

```python
retailer_sales_by_region  = df.groupby(['Retailer', 'Region'])['Total Sales'].sum()

retailer_sales_by_region_df = retailer_sales_by_region.unstack(level=1)

# Mostrar el DataFrame resultante para verificar su estructura
retailer_sales_by_region_df

# Crear un gr谩fico de barras para visualizar las ventas totales por minorista y regi贸n
retailer_sales_by_region_df.plot(kind='bar', title='Total Sales by Retailer and Region', figsize=(12, 6))
```

```python
# Agrupar los datos por a帽o y mes, y calcular la suma de las ventas totales para cada grupo
df_grouped = df.groupby(['year', 'month']) 

# Calcular la suma de las ventas totales para cada grupo de a帽o y mes
sales_by_year_month = df_grouped["Total Sales"].sum()

# Crear un gr谩fico de l铆neas para visualizar las ventas totales por a帽o y mes
sales_by_year_month.plot(kind='line', title='Total Sales by Year and Month', figsize=(12, 6))

```

```python
# Agrupar los datos por a帽o y mes, y calcular la suma de las unidades vendidas para cada grupo
units_by_year_month = df_grouped['Units Sold'].sum()

# Crear un gr谩fico de l铆neas para visualizar las unidades vendidas por a帽o y mes
units_by_year_month.plot(kind='line', title='Units Sold by Year and Month', figsize=(12, 6))
```

```python
import plotly.express as px
import pandas as pd

# Agrupar los datos por regi贸n, estado y ciudad, y calcular la suma de las ventas totales para cada grupo
df_grouped = df.groupby(['Region', 'State', 'City'])

# Calcular la suma de las ventas totales para cada grupo de regi贸n, estado y ciudad
sales_by_location = df_grouped['Total Sales'].sum()
units_by_location = df_grouped['Units Sold'].sum()

print(units_by_location)

summary_df = pd.DataFrame({
    'Region': sales_by_location.index.get_level_values('Region'),
    'State': sales_by_location.index.get_level_values('State'),
    'City': sales_by_location.index.get_level_values('City'),
    'Total Sales': sales_by_location.values,
    'Units Sold': units_by_location.values
})

fig = px.treemap(summary_df,
                 path=['Region', 'State', 'City'],
                 values='Total Sales',
                 color='Units Sold',
                 title='Distribuci贸n de Ventas por Ubicaci贸n')
fig.show()
```
