# Definir Listas y tuplas

```python
# lista de libros
libros = ["Cien Años de Soledad", 
          "Don Quijote de la Mancha", 
          "La Sombra del Viento", 
          "El Amor en los Tiempos del Cólera", 
          "Rayuela",
          "Ficciones"]

# tupla de generos literarios
generos = ("Ficción", "Misterio", "Romance", "Ciencia", "Historia")

# Lista del Estado del libro
estados = ["Disponible"] * len(libros)
```

# Verificaion rapida

```python

print("libros", libros)
print("Géneros:", generos)
print("Estados:", estados)
print("Cantidad libros:", len(libros), "Cantidad estados:", len(estados))
```

## Recorrer los libros y mostrar estado 

```python
for i in range(len(libros)):
    print(f"Libro #{i+1}: {libros[i]} - Estado: {estados[i]}" )
```

## Asignar un genero aletoria a un libro y mostrarlos

```python
import random

print("Asignar un genero aletoria a un libro")

generos_asignados = []

# Asignar aleatoriamente el genero
for i in range(len(libros)):
    genero = random.choice(generos)
    generos_asignados.append(genero)

# Mostrar los libros con los generos asignados

for i in range(len(libros)):
    print(f"Libro #{i+1}: {libros[i]} - Genero: {generos_asignados[i]}" )
```

## Simular el préstamo de 3 libros (Disponible → Prestado)

```python
prestados = 0
indice = 0

while prestados < 3 and indice < len(estados):
    if estados[indice] == "Disponible":
        estados[indice] = "Prestado"
        prestados += 1
    indice += 1


for i in range(len(libros)):
    print(f"{libros[i]} - Estado: {estados[i]}")
```

## Contar cuántos libros están disponibles después de los préstamos

```python
disponibles  = 0

indice = 0

while indice < len(estados):
    
    if estados[indice] == "Disponible":
    
        disponibles += 1

    indice += 1    

print(f"Cantidad de libros prestados: {disponibles}")    
```
