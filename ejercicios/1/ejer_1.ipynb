# Definir Listas y tuplas

```python
# lista de libros
libros = ["Cien A√±os de Soledad", 
          "Don Quijote de la Mancha", 
          "La Sombra del Viento", 
          "El Amor en los Tiempos del C√≥lera", 
          "Rayuela",
          "Ficciones"]

# tupla de generos literarios
generos = ("Ficci√≥n", "Misterio", "Romance", "Ciencia", "Historia")

# Lista del Estado del libro
estados = ["Disponible"] * len(libros)
```

# Verificaion rapida

```python

print("libros", libros)
print("G√©neros:", generos)
print("Estados:", estados)
print("Cantidad libros:", len(libros), "Cantidad estados:", len(estados))
```

## Recorrer los libros y mostrar estado 

```python
for i in range(len(libros)):
    print(f"Libro #{i+1}: {libros[i]} - Estado: {estados[i]}" )
```

## Asignar un genero aletoria a un libro y mostrarlos

```python
import random

print("Asignar un genero aletoria a un libro")

generos_asignados = []

# Asignar aleatoriamente el genero
for i in range(len(libros)):
    genero = random.choice(generos)
    generos_asignados.append(genero)

# Mostrar los libros con los generos asignados

for i in range(len(libros)):
    print(f"Libro #{i+1}: {libros[i]} - Genero: {generos_asignados[i]}" )
```

## Simular el pr√©stamo de 3 libros (Disponible ‚Üí Prestado)

```python
prestados = 0
indice = 0

while prestados < 3 and indice < len(estados):
    if estados[indice] == "Disponible":
        estados[indice] = "Prestado"
        prestados += 1
    indice += 1


for i in range(len(libros)):
    print(f"{libros[i]} - Estado: {estados[i]}")
```

## Contar cu√°ntos libros est√°n disponibles despu√©s de los pr√©stamos

```python
disponibles  = 0

indice = 0

while indice < len(estados):
    
    if estados[indice] == "Disponible":
    
        disponibles += 1

    indice += 1    

print(f"Cantidad de libros prestados: {disponibles}")    
```

### Reporte de libros por genero

```python
# PARTE 2D.1: Reporte de libros por g√©nero

reporte_generos = {}

for genero in generos:
    reporte_generos[genero] = 0  # inicia conteo en 0

for i in range(len(generos_asignados)):
    genero_libro = generos_asignados[i]
    reporte_generos[genero_libro] += 1

print("\nREPORTE DE LIBROS POR G√âNERO:")
for genero in generos:
    print(f"{genero}: {reporte_generos[genero]} libro(s)")
```

### Encontrar el primer libro disponible de g√©nero "Ficci√≥n

```python
i = 0
encontrado = False

while i < len(libros)  and not encontrado:
    
    if generos_asignados[i] == "Ficc√≥n" and estados[i] == "Disponible":
        print("\n Primer Libro disponible de 'FICCI√ìN':")
        print(f"{libros[i]} (Estado: {estados[i]})")
        encontrado = True
    
    i += 1    

if not encontrado:
    print("No se encontro ningun libro")    
```

### Sistema de Devolucion de libros

```python
libro_devolver = input("Introduce el titulo del libro a devolver: ")
encontrado = False

for i in range(len(libros)):
    if libros[i] == libro_devolver and estados[i] == "Prestado":
        estados[i] = "Disponible"
        print(f"{libros[i]} ha sido devuelto y ahora est√° disponible.")
        encontrado = True
        break

if not encontrado:
    print("No se encontr√≥ el libro a devolver o no estaba prestado.")
```

## Estadistica de la bibloteca 

```python

total = len(libros)

disponibles = 0

prestados = 0

i = 0

while i < len(estados):
    if estados[i] == "Disponible":
        disponibles += 1
    else:
        prestados += 1
    i += 1

porcentaje_disp = (disponibles / total) * 100
porcentaje_pres = (prestados / total) * 100   

# G√©nero m√°s frecuente (usando el reporte ya hecho o rehaci√©ndolo)

conteo_generos = {}

# Incializamos el conteo en cero
for g in generos:
    conteo_generos[g] = 0

# Contamos por genero
for i in range(len(generos_asignados)):
    conteo_generos[generos_asignados[i]] += 1

# Encontrar el g√©nero con mayor cantidad

genero_top = None # Variable para almacenar el g√©nero m√°s frecuente
max_cant = -1 # Variable para almacenar la cantidad m√°xima encontrada (iniciada en -1 para asegurar que cualquier conteo ser√° mayor)

# Por que -1 y no 0?
# Se inicializa max_cant en -1 para garantizar que cualquier cantidad v√°lida de libros (mayor o igual a 0) pueda reemplazarlo en la primera iteraci√≥n, evitando errores en casos donde todos los valores sean cero.

for g in generos:
    if conteo_generos[g] > max_cant:
        max_cant = conteo_generos[g]
        genero_top = g 


print("\nüìä ESTAD√çSTICAS DE LA BIBLIOTECA")
print(f"Total de libros: {total}")
print(f"Disponibles: {disponibles} ({porcentaje_disp:.2f}%)")
print(f"Prestados: {prestados} ({porcentaje_pres:.2f}%)")
print(f"G√©nero m√°s frecuente: {genero_top} ({max_cant} libro(s))")
```
